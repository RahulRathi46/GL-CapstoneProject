```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sys import path as spath
from os import path as opath
spath.append(opath.join(opath.dirname('__file__'), '..', 'scripts'))
import Profiler as pf
from sklearn.model_selection import cross_val_score , KFold
```

1. Using Filled_Zero Dataset

```python
df = pd.read_csv('../dataset/filled_0/merged.csv' , index_col=[0])
```

```python
df.head()
```

```python
df['Type'] = df['Type'].map(
    {
        'A' : 0,
        'B' : 1,
        'C' : 2
    }
)
```

```python
x = df.drop(['Date', 'Weekly_Sales'] , axis = 1)
y = df['Weekly_Sales']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x , y, test_size=.30, random_state=0)
```

```python
pf.Base_Model_Profile(xtrain, xtest, ytrain, ytest)
```

As DecisionTreeRegressor Giving Best Accuracy , Now Testing The Dataset With CV.

```python
from sklearn.tree import DecisionTreeRegressor
scores = cross_val_score(DecisionTreeRegressor(), x , y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
kf = KFold(n_splits=10)
scores = cross_val_score(DecisionTreeRegressor(), x , y, cv=kf)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

3. Testing Filled Zero Scaled Dataset and Cross Vaildate

```python
df = pd.read_csv('../dataset/filled_0/merged.csv' , index_col=[0])
```

```python
df.head()
```

```python
df['Type'] = df['Type'].map(
    {
        'A' : 0,
        'B' : 1,
        'C' : 2
    }
)
```

```python
x = df.drop(['Date', 'Weekly_Sales'] , axis = 1)
y = df['Weekly_Sales']
```

```python
from sklearn.preprocessing import StandardScaler
cols = x.select_dtypes(['float' , 'int']).columns 
x.loc[: , cols] = StandardScaler().fit_transform(x.loc[: , cols])
```

```python
x.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x , y, test_size=.30, random_state=0)
```

```python
pf.Base_Model_Profile(xtrain, xtest, ytrain, ytest)
```

As DecisionTreeRegressor Giving Best Accuracy , Now Testing The Dataset With CV.

```python
scores = cross_val_score(DecisionTreeRegressor(), x , y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
kf = KFold(n_splits=10)
scores = cross_val_score(DecisionTreeRegressor(), x , y, cv=kf)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

There is no sifiginifac with scaled data and non scaled data with kfold validation.
